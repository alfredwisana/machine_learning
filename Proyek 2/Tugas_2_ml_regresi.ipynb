```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv(r"C:\Users\LENOVO\Documents\school\SEM 5\Machine Learning\CarPrice_Assignment.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df.drop(columns=['fueltype','aspiration','doornumber','carbody','drivewheel',
                 'enginelocation','wheelbase','CarName','enginetype',
                 'cylindernumber','fuelsystem'],inplace=True)
df.head()
```

```python
df.corr()
```

```python
df.drop(columns=['symboling','carheight','highwaympg','citympg','peakrpm','stroke'],inplace=True)
df.head()
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
y=y.values.reshape(-1,1)
y.shape
```

```python
scaling=MinMaxScaler()
X=scaling.fit_transform(X)
y=scaling.fit_transform(y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=21)
```

```python
X_train.shape 
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
print("Num GPUs Available: ", len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

```python
dnn = Sequential([
    Dense(units=64, input_shape=(8,), activation='relu'),
    Dense(units=32, activation='relu'),
    Dense(units=1, activation='linear')
])
```

```python
dnn.summary()
```

```python
dnn.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error', metrics=['accuracy'])
```

```python
dnn.fit(x=X_train,y=y_train, validation_split=0.01, epochs=50, batch_size=16, validation_data=(X_test,y_test))
```

```python
y_pred = dnn.predict(X_test)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```
